```python
import pandas as pd
import random
random.seed(42)
# First, count total number of rows
with open("./project/accepted_2007_to_2018Q4.csv") as f:
    total_lines = sum(1 for line in f) - 1  # exclude header

n_skip = total_lines - 5000
skip = sorted(random.sample(range(1, total_lines + 1), total_lines - 5000))

cols = [
    'int_rate',
    'dti',
    'loan_amnt',
    'installment',
    'annual_inc',
    'loan_status'
]

df_sample = pd.read_csv(
    "./project/accepted_2007_to_2018Q4.csv",
    usecols=cols,
    skiprows=skip  # skip most lines, keep only 5,000
)
df_sample['default'] = df_sample['loan_status'].apply(lambda x: 0 if 'Fully Paid' in str(x) else 1)

for col in cols:
    if col != 'default' and col != 'loan_status':
        df_sample[f's_{col}'] = (df_sample[col] - df_sample[col].mean()) / df_sample[col].std()

ordered_cols = []
for col in cols:
    if col != 'default' and col != 'loan_status':
        ordered_cols.append(col)
        ordered_cols.append(f's_{col}')
ordered_cols.append('default')  # add 'default' last

# Reorder the DataFrame
df_sample = df_sample[ordered_cols]
df_sample.to_csv("./project/processed_data.csv", index=False)

s_cols = [
    's_int_rate',
    's_dti',
    's_loan_amnt',
    's_installment',
    's_annual_inc',
    'default'
]
df_sample = df_sample[s_cols]
df_sample.to_csv("./project/processed_data_standarized_only.csv", index=False)
```
